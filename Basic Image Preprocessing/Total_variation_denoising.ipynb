## Denoising using Total Variation method

### Simple image

```python
import numpy as np
from skimage.restoration import denoise_tv_chambolle
import matplotlib.pyplot as plt

# Generate a noisy image
np.random.seed(42)
clean_image = np.zeros((100, 100))
clean_image[20:80, 20:80] = 1.0  # Create a square in the image
noisy_image = clean_image + 0.1 * np.random.randn(*clean_image.shape)

# Denoise using total variation denoising
denoised_image_tv = denoise_tv_chambolle(noisy_image, weight=0.9)

# Plot the results
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.title('Clean Image')
plt.imshow(clean_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Noisy Image')
plt.imshow(noisy_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Denoised Image (Total Variation)')
plt.imshow(denoised_image_tv, cmap='gray')
plt.axis('off')

plt.show()
```

### Brain image

```python
import numpy as np
import cv2
from skimage.restoration import denoise_tv_chambolle
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('brain_mri.png') 

# Add Gaussian noise to the image
np.random.seed(42)
noise_factor = 100
noise = noise_factor * np.random.randn(*image.shape)
noisy_brain_image = np.clip(image + noise, 0, 255).astype(np.uint8)

# Denoise using total variation denoising
denoised_brain_image_tv = denoise_tv_chambolle(noisy_brain_image, weight=0.1)

# Plot the results
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.title('Clean Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Noisy Image')
plt.imshow(noisy_brain_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Denoised Image (Total Variation)')
plt.imshow(denoised_brain_image_tv, cmap='gray')
plt.axis('off')

plt.show()
```
