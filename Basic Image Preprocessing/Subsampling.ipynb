## Subsampling an Image

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('brain_mri.png')  

# Subsample by removing every other row and column
subsampled_image = image[::32, ::32]

# Display the original and subsampled images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(subsampled_image)
plt.title('Subsampled Image')

plt.show()

# Save the image
plt.imsave('subsampled_image.png', subsampled_image, cmap='gray')
```

### Unsampling an image

```python
# Load the image
subsampled_image = cv2.imread('subsampled_image.png')

# Display the original image
plt.imshow(subsampled_image)
plt.title('subsampled_image')
plt.axis('off')
plt.show()

# Upsample the image (e.g., by a factor of 2)
upsampled_image = cv2.resize(subsampled_image, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)

# Display the upsampled image
plt.imshow(upsampled_image)
plt.title('Upsampled Image')
plt.axis('off')
plt.show()
```
