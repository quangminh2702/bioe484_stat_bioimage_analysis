## Finding Fourier Transform of Functions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

```python
# Define the time-domain signal
def x(t):

    return np.sin(2 * np.pi * t)

# Generate time values
t = np.linspace(-5, 5, 1000, endpoint=False)

# Calculate the signal values
x_t = x(t)
```

```python
# Calculate the Fourier Transform
frequencies = fftfreq(len(t), t[1] - t[0])
X_f = fft(x_t)/len(t)  #normalized wrt the length of data

# Plot the time-domain signal
plt.subplot(2, 1, 1)
plt.plot(t, x_t)
plt.title('Time-Domain Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Plot the magnitude of the Fourier Transform
plt.subplot(2, 1, 2)
plt.plot(frequencies, np.abs(X_f))
plt.title('Fourier Transform')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.xlim(-40,40)

plt.tight_layout()
plt.show()
```

```python
# Calculate the magnitude
magnitude = np.abs(X_f)
# Remove elements below the threshold (0.1)
threshold = 0.01
a = magnitude[magnitude > threshold]

print(a)
```
