```python
import numpy as np
import matplotlib.pyplot as plt

def sinusoid(t):
    return np.sin(2 * np.pi * t)

# Generate time values
t_continuous = np.linspace(-5, 5, 1000, endpoint=False)
t_sampled = np.arange(-5, 5, 0.1)  # Sample every 0.1 seconds

# Calculate the sample values
x_t_sampled = sinusoid(t_sampled)

plt.subplot(2, 1, 1)
plt.plot(t_continuous, sinusoid(t_continuous), label='Original Sinusoid')
plt.title('Original Sinusoid Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.stem(t_sampled, x_t_sampled, label='Sampled data', use_line_collection=True)
plt.title('Sampled data')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
