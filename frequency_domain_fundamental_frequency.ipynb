```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Define the time-domain signal
def x(t):
    return np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)

# Generate time values
t = np.linspace(0, 1, 1000, endpoint=False)

# Calculate the signal values
x_t = x(t)

# Calculate the Fourier Transform
frequencies = fftfreq(len(t), t[1] - t[0])
X_f = fft(x_t)

# Plot the magnitude spectrum
plt.plot(frequencies, np.abs(X_f))
plt.title('Magnitude Spectrum')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.show()

# Find the fundamental frequency
fundamental_frequency = frequencies[np.argmax(np.abs(X_f))]
print('Fundamental Frequency:', fundamental_frequency)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time-domain signal
def x(t):
    return np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)

# Generate time values
t = np.linspace(0, 1, 1000, endpoint=False)

# Calculate Fourier coefficients
a_0 = np.mean(x(t))
a_n = 2 * np.abs(np.fft.fft(x(t))) / len(t)

# Express the function as a sum of sinusoidal functions
def sum_of_sinusoids(t, a_0, a_n, frequencies):
    result = a_0 * np.ones_like(t)  # Initialize with DC component
    for n, freq in enumerate(frequencies, start=1):
        result += a_n[n] * np.cos(2 * np.pi * freq * t + np.angle(np.fft.fft(x(t))[n]))
    return result

# Plot the original function
plt.figure(figsize=(12, 6))
plt.plot(t, x(t), label='Original Function')

# Plot the sum of sinusoidal functions
frequencies = [5, 10]  # Fundamental frequency and harmonics
result = sum_of_sinusoids(t, a_0, a_n, frequencies)
plt.plot(t, result, label='Sum of Sinusoidal Functions', linestyle='--')

plt.title('Original Function and Sum of Sinusoidal Functions')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()

```
